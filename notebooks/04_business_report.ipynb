# 1. Key Findings

Bu çalışma kapsamında havayolu yolcu memnuniyetini etkileyen faktörler, veri temizleme ve keşifsel veri analizi (EDA) yöntemleri kullanılarak incelenmiştir. Yapılan analizler sonucunda müşteri memnuniyetini etkileyen temel değişkenler belirlenmiş ve aşağıdaki önemli bulgular elde edilmiştir.

### Başlıca Bulgular

- Veri setinde yalnızca **Arrival Delay in Minutes** değişkeninde eksik değer tespit edilmiş, bu değerler **Departure Delay in Minutes** değişkeni kullanılarak doldurulmuştur.

- Yolcuların yaklaşık **%82'si Loyal Customer**, **%18'i ise Disloyal Customer** grubunda yer almaktadır.

- Yolcuların yaklaşık **%69'u Business Travel**, **%31'i Personal Travel** amacıyla seyahat etmektedir.

- İncelenen yolcuların yaklaşık **%57'si memnun değildir**, bu durum hizmet kalitesinin geliştirilmesi için önemli bir iyileştirme alanı olduğunu göstermektedir.

- **Business Class** yolcularının memnuniyet oranı Economy ve Eco Plus sınıflarına kıyasla belirgin şekilde daha yüksektir.

- **Loyal Customer** grubunda memnuniyet oranı, Disloyal Customer grubuna göre daha yüksektir.

- **Business Travel** yapan yolcuların memnuniyet düzeyi, Personal Travel yapan yolculara göre belirgin şekilde yüksektir.

- **Gender** değişkeni incelendiğinde kadın ve erkek yolcular arasında memnuniyet açısından belirgin bir farklılık gözlemlenmemiştir.

- Uçuş mesafesi arttıkça müşteri memnuniyetinin artma eğiliminde olduğu görülmüştür.

- Kalkış ve varış gecikmeleri arttıkça müşteri memnuniyetinin azalma eğiliminde olduğu gözlemlenmiştir.

- Departure Delay ve Arrival Delay değişkenleri arasında güçlü pozitif doğrusal ilişki bulunmaktadır.

# 1. Executive Summary

Bu çalışma, havayolu yolcu memnuniyetini etkileyen temel faktörleri belirlemek amacıyla gerçekleştirilmiştir. Çalışma kapsamında veri seti üzerinde veri temizleme, keşifsel veri analizi (Exploratory Data Analysis - EDA) ve görselleştirme çalışmaları yürütülmüştür.

Analizler sonucunda müşteri memnuniyetinin; uçuş sınıfı, müşteri tipi, seyahat amacı, uçuş mesafesi ve uçuş gecikmeleri gibi operasyonel ve hizmet odaklı değişkenlerden önemli ölçüde etkilendiği görülmüştür. Buna karşılık cinsiyet değişkeninin müşteri memnuniyeti üzerinde belirgin bir etkisi gözlemlenmemiştir.

Elde edilen bulgular doğrultusunda, havayolu şirketlerinin müşteri deneyimini iyileştirmesine katkı sağlayabilecek iş odaklı öneriler geliştirilmiştir.

## 2. Key Findings

### 1. Uçuş sınıfı müşteri memnuniyetini belirleyen en önemli faktörlerden biridir.

### 1. Business Class yolcuları en yüksek memnuniyet oranına sahiptir.

Business Class'ta seyahat eden yolcuların yaklaşık **%69'u** memnun olduğunu belirtirken, bu oran Economy sınıfında **%19**, Eco Plus sınıfında ise **%25** seviyesindedir. Bu bulgu, sunulan hizmet kalitesinin ve uçuş deneyiminin müşteri memnuniyeti üzerinde güçlü bir etkiye sahip olduğunu göstermektedir.

## 2. Key Findings

### 2.1 Flight Class is the Strongest Indicator of Customer Satisfaction

Business Class yolcularının yaklaşık **%69'u** memnun olduğunu belirtirken, bu oran Economy sınıfında yalnızca **%19**, Eco Plus sınıfında ise **%25** seviyesindedir.

Bu sonuç, müşteri memnuniyetinin yalnızca uçuşun gerçekleşmesiyle değil, sunulan hizmet kalitesi ve yolculuk deneyimiyle de güçlü biçimde ilişkili olduğunu göstermektedir.

### 2.2 Travel Purpose Has a Significant Impact on Customer Satisfaction

Business Travel yapan yolcuların yaklaşık **%58'i** memnun olduğunu belirtirken, Personal Travel grubunda bu oran yalnızca **%10** seviyesindedir.

Bu durum, iş amaçlı seyahat eden yolcuların sunulan hizmetlerden daha fazla fayda sağladığını veya beklentilerinin mevcut hizmetlerle daha iyi karşılandığını düşündürmektedir.

### 2.3 Loyal Customers Report Higher Satisfaction Levels

Sadık müşterilerin memnuniyet oranı, sadık olmayan müşterilere göre belirgin şekilde daha yüksektir. Buna rağmen sadık müşterilerin yaklaşık yarısının hâlâ memnun olmadığını belirtmesi, mevcut sadakat programlarının geliştirilmesi için önemli bir fırsat sunmaktadır.

### 2.4 Uçuş Gecikmeleri Müşteri Memnuniyetini Olumsuz Etkilemektedir

Analiz sonuçları, hem kalkış hem de varış gecikmelerinin memnun olmayan yolcularda daha yüksek seviyelerde gerçekleştiğini göstermektedir. Özellikle memnun olmayan yolcuların ortalama gecikme süreleri, memnun olan yolculara kıyasla daha yüksektir.

Bu bulgu, operasyonel süreçlerde yaşanan gecikmelerin müşteri deneyimini olumsuz etkileyen önemli faktörlerden biri olduğunu göstermektedir. Gecikmelerin azaltılmasına yönelik operasyonel iyileştirmeler, müşteri memnuniyetinin artırılmasına katkı sağlayabilir.

# 3. Business Recommendations

Veri analizi sonucunda elde edilen bulgular doğrultusunda, müşteri memnuniyetini artırmaya yönelik uygulanabilir iş önerileri geliştirilmiştir. Aşağıdaki öneriler, analizlerde tespit edilen eğilimlere dayanmakta olup müşteri deneyiminin iyileştirilmesine katkı sağlayabilecek stratejik aksiyonları içermektedir.

### 3.1 Economy Sınıfındaki Yolcu Deneyimi İyileştirilmelidir

Analizler, Economy sınıfında seyahat eden yolcuların memnuniyet oranının Business Class yolcularına kıyasla oldukça düşük olduğunu göstermektedir.

Bu nedenle özellikle Economy sınıfında sunulan hizmetlerin geliştirilmesi öncelikli iyileştirme alanı olarak değerlendirilmelidir.

Özellikle;

- Koltuk konforunun artırılması,
- Uçuş içi eğlence hizmetlerinin geliştirilmesi,
- Yiyecek ve içecek kalitesinin iyileştirilmesi,
- Kabin hizmetlerinin güçlendirilmesi

müşteri memnuniyetinin artırılmasına katkı sağlayabilir.

### 3.2 Operasyonel Gecikmeler Azaltılmalıdır

Analiz sonuçları, kalkış ve varış gecikmeleri arttıkça müşteri memnuniyetinin azalma eğiliminde olduğunu göstermektedir.

Bu nedenle operasyonel süreçlerde yaşanan gecikmelerin azaltılması müşteri deneyimini iyileştirmek açısından kritik öneme sahiptir.

Özellikle;

- Yer hizmetleri süreçlerinin optimize edilmesi,
- Uçuş planlamasının iyileştirilmesi,
- Kapı değişikliklerinin azaltılması,
- Gecikme yaşandığında yolcuların zamanında bilgilendirilmesi

müşteri memnuniyetini olumlu yönde etkileyebilir.

### 3.3 Sadakat Programları Güçlendirilmelidir

Sadık müşterilerin memnuniyet düzeyi daha yüksek olsa da bu grubun önemli bir kısmının hâlâ memnun olmadığını belirtmesi dikkat çekmektedir.

Bu nedenle mevcut sadakat programlarının yalnızca puan veya mil kazandıran sistemler olmaktan çıkarılarak müşteri deneyimini artıracak ayrıcalıklarla desteklenmesi önerilmektedir.

Örneğin;

- Öncelikli check-in,
- Ücretsiz koltuk seçimi,
- Lounge erişimi,
- Kişiselleştirilmiş kampanyalar

gibi uygulamalar müşteri bağlılığını artırabilir.

### 3.4 Personal Travel Yolcularına Yönelik Hizmetler Geliştirilmelidir

Analizler, kişisel amaçlı seyahat eden yolcuların memnuniyet oranının iş amaçlı seyahat eden yolculara göre oldukça düşük olduğunu göstermektedir.

Bu nedenle tatil veya bireysel seyahat eden yolcuların beklentilerini karşılayacak hizmetlerin geliştirilmesi önem taşımaktadır.

Özellikle ailelere yönelik kampanyalar, eğlence hizmetleri, esnek bilet seçenekleri ve kişiselleştirilmiş hizmetler bu müşteri grubunun memnuniyetini artırabilir.

# 4. Conclusion

Bu çalışma kapsamında havayolu yolcu memnuniyetini etkileyen temel faktörler veri analizi yöntemleri kullanılarak incelenmiştir. Yapılan analizler sonucunda müşteri memnuniyetinin özellikle uçuş sınıfı, seyahat amacı, müşteri tipi ve operasyonel gecikmelerden önemli ölçüde etkilendiği görülmüştür.

Elde edilen bulgular doğrultusunda geliştirilen öneriler, havayolu şirketlerinin müşteri deneyimini iyileştirmesine ve hizmet kalitesini artırmasına katkı sağlayabilecek niteliktedir.

Bu çalışma, veri analizi süreçlerinin yalnızca mevcut durumu açıklamakla kalmayıp, iş kararlarını destekleyen stratejik içgörüler sunabileceğini göstermektedir.
